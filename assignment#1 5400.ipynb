```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
res = requests.get('https://www.trustpilot.com/review/residentialadvance.com')
res.status_code
```

```python
html = res.text
soup = BeautifulSoup(html)
```

```python
review_content = soup.find_all(name='section', attrs={'class':'styles_reviewContentwrapper__zH_9M'})
```

```python
review_scrapper = []

page_url = 'https://www.trustpilot.com/review/listwithclever.com'

while len(review_scrapper) < 600:
    res = requests.get(page_url)
    html = res.text
    soup = BeautifulSoup(html)
    
    review_content = soup.find_all(name='section', attrs={'class': 'styles_reviewContentwrapper__zH_9M'})
    
    for review in review_content:
        try:
            companyName = 'Clever Real Estate '
            
            dateElement = review.find(name='time', attrs={'data-service-review-date-time-ago': "true"})
            datePublished = dateElement.text if dateElement else "N/A"
            
            ratingElement = review.find(name='div', attrs={'class': 'styles_reviewHeader__iU9Px'})
            ratingValue = ratingElement.attrs['data-service-review-rating'] if ratingElement else "N/A"
            
            reviewBodyElement = review.find(name='p', attrs={'class': 'typography_body-l__KUYFJ typography_appearance-default__AAY17 typography_color-black__5LYEn'})
            reviewBody = reviewBodyElement.text if reviewBodyElement else "N/A"
            
            review_scrapper.append([companyName, datePublished, ratingValue, reviewBody])
        
        except Exception as e:
            print(f"Error while processing review: {e}")
    
    # Iterate over each page
    page = soup.find(name='a', attrs={'class': 'link_internal__7XN06 button_button__T34Lr button_m__lq0nA button_appearance-outline__vYcdF button_squared__21GoE link_button___108l pagination-link_next__SDNU4 pagination-link_rel__VElFy'})
    if page:
        page_url = 'https://www.trustpilot.com' + page.attrs['href']
    else:
        break

review_scrapper
```

```python
with open('review_scrapper.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f, delimiter=',')
    writer.writerow(['companyName', 'datePublished', 'ratingValue', 'reviewBody'])
    writer.writerows(review_scrapper)
```
